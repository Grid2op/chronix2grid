```python
import pandas as pd
import os
```

```python
env_name = "case118_l2rpn_icaps_2.5x"
path_ref = os.path.join("..", "example", "input", "generation")
input_path = os.path.join(path_ref, env_name)
df = pd.read_csv(os.path.join(input_path, "prods_charac.csv"))

avg_pmaxs = df.groupby(["type"])["Pmax"].mean()
types = avg_pmaxs.index.to_numpy()
```

```python
df
```

```python
mask = (df["type"] == "wind") | (df["type"] == "solar")
df.at[mask, "Pmax"] = df[mask]["Pmax"] * 1.5
```

```python
mask = df["type"] == "thermal"
df.at[mask, "start_cost"] = df[mask]["start_cost"] * 10
```

```python
path_tmp = os.path.join("..", "example", "custom", "input", "generation")
input_path = os.path.join(path_tmp, env_name)
df.to_csv(os.path.join(input_path, "prods_charac.csv"), index=False)
```
